```python
df= pd.read_csv("CountryTable.csv", index_col=2)
df
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion
filtro6=df[df["gnp"] > 2006]
filtro6
```

**FILTROS BASICOS CON CONETORES LOGICOS**

```python
#Filtro Y
filtro7=df[(df["population"] > 900) & (df["life_expectancy"] == "Republic")]
filtro7
```

```python
#Filtro o
filtro8=df[(df["population"] > 900) | (df["life_expectancy"] == "Republic")]
filtro8
```

```python
#Filtro not
filtro9=df[~(df["population"] >= 900) & ~(df["life_expectancy"] == "Republic")]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("Actividad1RoyMata.csv")
```
